## **This dataset is having 45,100 results of international football matches starting from the very first official match in 1872 up to 2023.**

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
file = "./results.csv"
data = pd.read_csv(file)
data.head(5)
```

## **This dataset has 9 columns**

```python
data.shape
```

```python
data.isnull().sum()
```

## **Convert the datatype of data in date column into datetime**

```python
print(data.dtypes)
```

```python
data['date'] = pd.to_datetime(data['date'])
print(data.dtypes)
```

```python
from datetime import datetime
```

## **Now, I will filter the list to know the number of Japan national team's home games and wins**

```python
home = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1)) & (data['home_team'] == 'Japan')]
home.shape
```

```python
home_wins = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['home_team'] == 'Japan') & (data['home_score'] > data['away_score'])]
home_wins.shape
```

```python
home_draws = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['home_team'] == 'Japan') & (data['home_score'] == data['away_score'])]
home_draws.shape
```

```python
home_loses = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['home_team'] == 'Japan') & (data['home_score'] < data['away_score'])]
home_loses.shape
```

```python
home_wins_num = home_wins["home_team"].count()
home_draws_num = home_draws["home_team"].count()
home_loses_num = home_loses["home_team"].count()

t = [home_wins_num, home_draws_num, home_loses_num]
colors = ["blue", "yelow", "red"]
labels = ["Win", "Draw", "Lose"]
plt.title('Each Percentage of National Japanse team win, draw and lose at home')
plt.pie(t, labels=labels, autopct='%1.1f%%')
plt.show()
```

## **At next, I will filter the list to know the number of Japan national team's away games and wins**

```python
away = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1)) & (data['away_team'] == 'Japan')]
away.shape
```

```python
away_wins = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['away_team'] == 'Japan') & (data['away_score'] > data['home_score'])]
away_wins.shape
```

```python
away_draws = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['away_team'] == 'Japan') & (data['away_score'] == data['home_score'])]
away_draws.shape
```

```python
away_loses = data[(data['date'] >= dt.datetime(1900,1,1)) & (data['date'] < dt.datetime(2000,1,1))
                  & (data['away_team'] == 'Japan') & (data['away_score'] < data['home_score'])]
away_loses.shape
```

```python
away_wins_num = away_wins["away_team"].count()
away_draws_num = away_draws["away_team"].count()
away_loses_num = away_loses["away_team"].count()

t = [away_wins_num, away_draws_num, away_loses_num]
colors = ["blue", "yelow", "red"]
labels = ["Win", "Draw", "Lose"]
plt.title('Each Percentage of National Japanse team win, draw and lose at away')
plt.pie(t, labels=labels, autopct='%1.1f%%')
plt.show()
```
